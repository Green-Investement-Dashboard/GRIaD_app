```python
import hashlib, binascii, os


"""Hash mot de passe SHA-256"""
password = 'test'
salt = hashlib.sha256(os.urandom(60)).hexdigest().encode('ascii')
pwdhash = hashlib.pbkdf2_hmac('sha512', password.encode('utf-8'), 
                            salt, 100000)
pwdhash = binascii.hexlify(pwdhash)
val = (salt + pwdhash) # return bytes

val
```

```python
df['password'] = val
df['two_fa_key'] = 'M5KQ74ZMSURPYF7H3NZBLIXRGF5AQ4JG'
df['two_fa'] = False
#df = df.drop(index=[2])
df=df.set_index('id')
df
```

```python
df.to_sql("User2", con, if_exists="replace")
```
